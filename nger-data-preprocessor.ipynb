```python
rainfall_filename = r'C:\Users\t5s\03-NGER-2019-20\04-Data\09. Raw data from BI\WRP_&_WWTP_Flow_and_Rainfall_Measures-_Date_Range_Report___4997986.xlsx'

rainfall_workbook = openpyxl.load_workbook(rainfall_filename)

all_flow_rate_data = list()

def find_row_for_total_flow(worksheet):
    # total is in column B (normally in row 373)
    cells_in_column_b = worksheet["b4:b400"]
    for row in cells_in_column_b:
            for cell in row:
                # check if this column has a known site reference
                if (cell.value == "Total Flow"):
                    return cell.row

for worksheet_name in rainfall_workbook.sheetnames:
        worksheet = rainfall_workbook[worksheet_name]
        
        row_number_for_totals = find_row_for_total_flow(worksheet)
        if row_number_for_totals is None:
            continue
        
        # the rainfall and flow sampling point reference is on row 4
        flow_reference_cells = worksheet["C4:Z4"]
        for column in flow_reference_cells:
            for cell in column:
                # check if this column has a known site reference
                if (cell.value != None and len(cell.value) > 2):
                    column_letter_for_sampling_point = cell.column_letter
                    sampling_point = cell.value
                    
                    cell_reference_for_total_flow = f"{column_letter_for_sampling_point}{row_number_for_totals}"
                    total_flow_for_sampling_point = worksheet[cell_reference_for_total_flow].value
                    total_flow_for_sampling_point = 0 if total_flow_for_sampling_point == None else total_flow_for_sampling_point
                    all_flow_rate_data.append({ "catchment": worksheet_name, "sampling_point": sampling_point, "total_flow_for_sampling_point": total_flow_for_sampling_point})

flow_rate_df = pd.DataFrame(all_flow_rate_data)
flow_rate_df.head()
```

```python
def get_data_for_analyte(data_for_site, analyte_name):
    analyte_data = data_for_site.query(f"`Analyte Name` == '{analyte_name}'")
    return analyte_data

workbook_filenames = [
    r'C:\Users\t5s\03-NGER-2019-20\06-Models\01. Models - Test\01.Bondi Facility WSAA - GHG and Energy Inventory Assessment (2019-20).xlsx',
    r'C:\Users\t5s\03-NGER-2019-20\06-Models\01. Models - Test\02.Cronulla Facility WSAA - GHG and Energy Inventory Assessment (2019-20).xlsx',
    r'C:\Users\t5s\03-NGER-2019-20\06-Models\01. Models - Test\03.Malabar Facility WSAA - GHG&Energy Inventory Assessment(2019-20).xlsx'
]

# Finds out which analytes are in use for a site in the WASAA model spreadsheet
# @param  column_letter_for_site  The column letter, e.g. D. Data for a site is found in a single column.
#                                 e.g. in the Bondi model, BN008 is in column D in worksheet 5A1. Bondi WWT (Data)
# @param  worksheet               The worksheet to look for data
# @return a list of dicts with keys analyte_name (the name of the analyte) and analyte_cell_range (the range where data
#         for that analyte should be pasted)
# @example [{ "analyte_name": "Total Nitrogen", "analyte_cell_range": "D19:D384" }]
def find_analytes(column_letter_for_site, worksheet):
    # We know that the analyte parameter name is in row 8 (for the first analyte)
    # and row 386 (for the second analyte)
    ROW_FOR_PARAMETER_NAME_ONE = 8
    ROW_FOR_PARAMETER_NAME_TWO = 386
    analyte_name_cells = [
        f"{column_letter_for_site}{ROW_FOR_PARAMETER_NAME_ONE}",
        f"{column_letter_for_site}{ROW_FOR_PARAMETER_NAME_TWO}"
    ]
    analytes = list()
    for analyte_name_cell in analyte_name_cells:
        cell = worksheet[analyte_name_cell]

        # the data starts 11 rows after the parameter name
        ANALYTE_DATA_ROW_START_OFFSET = 11
        TOTAL_NUM_ROWS_OF_DATA = 366
        ANALYTE_DATA_ROW_END_OFFSET = ANALYTE_DATA_ROW_START_OFFSET + TOTAL_NUM_ROWS_OF_DATA
        analyte_data_start_cell_reference = f"{cell.column_letter}{cell.row + ANALYTE_DATA_ROW_START_OFFSET}"
        analyte_data_end_cell_reference = f"{cell.column_letter}{cell.row + ANALYTE_DATA_ROW_END_OFFSET}"
        analytes.append({
            "analyte_name": cell.value,
            "analyte_cell_range": f"{analyte_data_start_cell_reference}:{analyte_data_end_cell_reference}"
        })
    
    return analytes

# Pastes analyte data for a site into the WASAA model spreadsheet
# @param  all_data  All the analyte data retrieved from BI
# @param  site_reference  e.g. BN0008
# @param  worksheet  The worksheet where data for that site belongs
# @param  column_letter_for_site  The column letter, e.g. D. Data for a site is found in a single column.
#                                 e.g. in the Bondi model, BN008 is in column D in worksheet 5A1. Bondi WWT (Data)
# @return void
def populate_data_for_model(all_data, site_reference, worksheet, column_letter_for_site):
    data_for_site = all_data.query(f"`Site Code` == '{site_reference}'")
  
    analytes = find_analytes(column_letter_for_site, worksheet)
    
    for analyte in analytes:
        analyte_name = analyte["analyte_name"]
        data_for_analyte = get_data_for_analyte(data_for_site, analyte_name)
        
        cell_range = worksheet[analyte["analyte_cell_range"]]
                
        idx = 0
        for row in cell_range:
            for cell in row:
                # check if there is still data - there might be less than 366 rows in the data
                if (idx < len(data_for_analyte['Result Numeric Value'])):
                    analyte_value_for_this_day = data_for_analyte['Result Numeric Value'].iloc[idx]
                    if (analyte_value_for_this_day != None and analyte_value_for_this_day != float('nan')):
                        cell.value = analyte_value_for_this_day
            idx = idx + 1
            
def populate_analyte_data(worksheet):
     # the site reference is on row 6
    site_reference_cells = worksheet["A6:AZ6"]
    for column in site_reference_cells:
        for cell in column:
            # check if this column has a known site reference
            if (cell.value != None and cell.value in all_site_references):
                print(f"Found site ref {cell.value} in worksheet {worksheet}. Processing...")
                column_letter_for_site = cell.column_letter
                site_reference = cell.value
                populate_data_for_model(all_data, site_reference, worksheet, column_letter_for_site)

all_flow_rate_sampling_point_references = flow_rate_df.sampling_point.unique()

from openpyxl.utils import get_column_letter

def find_flow_rate_location(worksheet, site_reference_cell):
    # this is to find the flow rate location so that data can be copied and pasted to WSAA model
    
    FLOW_DATA_ROW_START_OFFSET = 2
    TOTAL_NUM_ROWS_OF_FLOW_DATA = 4
    FLOW_DATA_ROW_END_OFFSET = FLOW_DATA_ROW_START_OFFSET + TOTAL_NUM_ROWS_OF_FLOW_DATA
    flow_data_start_cell_reference = f"{site_reference_cell.column_letter}{site_reference_cell.row + FLOW_DATA_ROW_START_OFFSET}"
    flow_data_end_cell_reference = f"{site_reference_cell.column_letter}{site_reference_cell.row + FLOW_DATA_ROW_END_OFFSET}"
        
    search_range = f"{flow_data_start_cell_reference}:{flow_data_end_cell_reference}"
    cells_in_search_rage_flow_rate = worksheet[search_range]
    for row in cells_in_search_rage_flow_rate:
            for cell in row:
                # check if this column has a known site reference
                if (cell.value == "Flow rate"):
                    # we put the flow rate 1 column across from the flow rate label
                    column_letter_for_flow_rate = get_column_letter(cell.column + 1)
                    return f"{column_letter_for_flow_rate}{cell.row}"

def populate_flow_rate_data(worksheet):
     # the sampling point reference is on row 17
    sampling_point_cells = worksheet["A16:AZ17"]
    for column in sampling_point_cells:
        for cell in column:
            # check if this column has a known site reference
            if (cell.value != None and str(cell.value) in all_flow_rate_sampling_point_references):
                sampling_point_ref = str(cell.value)
                print(f"Found sampling point ref {cell.value} in worksheet {worksheet}. Processing...")
                cell_reference_for_flow_Rate = find_flow_rate_location(worksheet, cell)
                print(cell_reference_for_flow_Rate)
                cell_for_flow_rate = worksheet[cell_reference_for_flow_Rate]
                
                print(f"sampling_point == {sampling_point_ref}")
                flow_rate_datum = flow_rate_df.query(f"sampling_point == '{sampling_point_ref}'")
                cell_for_flow_rate.value = flow_rate_datum['total_flow_for_sampling_point'].values[0]
                
for workbook_filename in workbook_filenames:
    print(f"processing workbook {workbook_filename}")
    workbook = openpyxl.load_workbook(workbook_filename)
    print("Loaded workbook")
    for worksheet_name in workbook.sheetnames:
        worksheet = workbook[worksheet_name]
        print(f"Processing worksheet {worksheet}")
#         populate_analyte_data(worksheet)
        print(worksheet)
    
        if ("PFD" in worksheet_name): 
            populate_flow_rate_data(worksheet)

    workbook.save(workbook_filename)
print("done")
```
